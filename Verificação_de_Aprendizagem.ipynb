```python
def soma(a, b):     
  return a+b

def multiplicacao(a,b):
  return a*b

def divisao(a,b):
  return a//b

def maior(a,b):
  if a > b:
    return f'{a} foi maior numero digitado'
  if b > a:
    return f'{b} foi o maior numero digitado'
  
  if a == b:
    return f'{a} = {b}'

def somaParOuImpar(a,b):
  if (a+b)%2 == 0:
    return 'A soma dos números é Par'
  else:
    return 'A soma dos números é Impar'

def maiorQue40(a,b):
  if multiplicacao(a, b) > 40:
    if divisao(a, b) != 0:
      return multiplicacao(a, b)/ divisao(a, b)
    else:
      return f'A parte inteira da fração {a}/{b} = 0'
  else:
    return multiplicacao(a, b)   

a = int(input('Digite o primeiro numero: ')) 
b = int(input('Digite o segundo numero: '))  
print('A soma dos numeros digitados é:', soma(a, b))
print('A multiplicação dos numeros digitados é:', multiplicacao(a, b))
print('A divisão inteira dos numeros digitados é:', divisao(a, b)) 
print(maior(a, b))
print(somaParOuImpar(a, b)) 
print(maiorQue40(a, b))
```

```python
fraseOriginal = input('Digite uma frase para serem contadas e retiradas suas vogais: ')
fraseSemVogal = ''
vogais = 0
for char in fraseOriginal:
  if char in ('a', 'e', 'i', 'o', 'u'):
    vogais += 1
  else:
    if len(fraseSemVogal) == 0 and char != ' ':
      fraseSemVogal += char
    elif len(fraseSemVogal) != 0:
      fraseSemVogal += char
print('A frase sem vogais fica:\n', fraseSemVogal)
print(vogais, 'vogais') 




```

```python
def fraseSemVogal(fraseOriginal): 
  fraseSemVogal = ''
  for char in fraseOriginal:
    if char not in ('a', 'e', 'i', 'o', 'u', 'ã', 'â', 'á', 'é', 'ê', 'í', 'ô', 'ó', 'ú'):
      if len(fraseSemVogal) == 0 and char != ' ':
        fraseSemVogal += char
      elif len(fraseSemVogal) != 0:
        fraseSemVogal += char
  return fraseSemVogal
def contaVogais(fraseOriginal):
  vogais = 0
  for char in fraseOriginal:
    if char in ('a', 'e', 'i', 'o', 'u', 'ã', 'â', 'á', 'é', 'ê', 'í', 'ô', 'ó', 'ú'): 
      vogais += 1
  return vogais
fraseOriginal = input('Digite uma frase para serem contadas e retiradas suas vogais: ')
print('A frase digitada sem vogais fica:\n', fraseSemVogal(fraseOriginal))
print('A frase digitada tem', contaVogais(fraseOriginal), 'vogais')




```

```python
answers = ['did you call the victim?','were you at the crime scene?',
             'do you live close to the victim', 'did you owe money to the victim',
             'have you wordked with the victim']

yesAccaunt = 0
noAccaunt = 0
result = ''

print('answer the questions whith [yes/no]: ')
for answer in answers:
  userAnswers = input(f'{answer}: ').lower()

# Accaunting
  if (userAnswers == 'yes'):
    yesAccaunt+=1
  elif (userAnswers == 'no'):
    noAccaunt+=1
  else:
    print('invalid response')
  
#evaluating suspicious
  if (yesAccaunt == 2):
    result = 'suspect'
  elif (yesAccaunt >= 3 and yesAccaunt <= 4):
    result = 'accomplice'
  elif (yesAccaunt == 5):
    result = 'assassin'

print()
print(f'positive: {yesAccaunt}')
print(f'negative: {noAccaunt}')
print(result)
```

```python
lista = list()
y = 0

for i in range(1,7+1):
      
  valores = int(input(f"Digite o {i}° valor: "))
  
  if valores % 2 == 0:
        
    y+=1
    lista.insert(0,valores)
    print(lista) 
    
  else:

    lista.insert(len(lista),valores)
    print(lista)
    
ordenado = sorted(lista[:y]) + sorted(lista[y:])

print()
print(ordenado)
```
